## Importing required packages


```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

## Geting car crashes dataset from seaborn

```python
df = pd.read_csv('bad-drivers.csv')
sns.set(rc={'figure.figsize':(20,8.27)})
df
```

```python
sns.barplot(x='State', y='Number of drivers involved in fatal collisions per billion miles' ,data=df)
```

```python
data_grouped_by_state = df.groupby('State', as_index=False).mean()
data_grouped_by_state.head()
```

## The relationship between number of Drivers Involved In Fatal Collisions Who Were Speeding and the amount of car insurance spend

```python
sns.relplot(x='Percentage Of Drivers Involved In Fatal Collisions Who Were Speeding', y='Car Insurance Premiums ($)', data=data_grouped_by_state , kind='line')
```

## The relationship between percentage of insurane losses and Percentage Of Drivers Involved In Fatal Collisions Who Were Alcohol-Impaired

```python
sns.boxplot(x='Percentage Of Drivers Involved In Fatal Collisions Who Were Alcohol-Impaired' ,y='Losses incurred by insurance companies for collisions per insured driver ($)', data=data_grouped_by_state)
```

## The relationship between Car Insurance Premiumsand osses incurred by insurance companies for collisions

```python
sns.relplot(x='Losses incurred by insurance companies for collisions per insured driver ($)', y='Car Insurance Premiums ($)', data=data_grouped_by_state , kind='line')
```

## The relationship betwwen the state and Percentage Of Drivers Involved In Fatal Collisions Who Were Speeding

```python
sns.barplot(x='State', y='Percentage Of Drivers Involved In Fatal Collisions Who Were Speeding' ,data=df)
```
